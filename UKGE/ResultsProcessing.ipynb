```python
import pandas as pd
import numpy as np
from polls import *
from UKGE import *
import ast
```

```python
df = pd.read_csv("outputs/constituencieswithresults.csv")
```

```python
df
```

```python
for c in ["2019_Results","2017_Results","2015_Results","2010_Results","2005_Results"]:
    for p in std_parties:
        df[c+p] = df[c].map(lambda x: ast.literal_eval(x)[p])
    df[c+"tot"] = df[[c+p for p in std_parties]].sum(axis=1)
    for p in std_parties:
        df[c+p+"_pc"] = df.apply(lambda x: ast.literal_eval(x[c])[p]/x[c+"tot"], axis=1)
df
```

```python
regdf = pd.read_excel("data/England-Parliamentary-electorates-for-2023.xlsx", sheet_name="CONSTITUENCIES", skiprows=0, skipfooter=57)
regdf = regdf.drop([0,1])
regdf["Constituency"] = regdf.apply(lambda x:x["CURRENT CONSTITUENCY ELECTORATE TABLE"].replace(" BC","").replace(" CC",""),axis=1)
regdf = regdf.loc[:,["Constituency","Region"]]
df["EngReg"] = df["New constituency name"].apply(lambda x: regdf[regdf["Constituency"]==x]["Region"].iloc[0] if x in list(regdf["Constituency"]) else 0)
df["Scot"] = df[df.columns[15]].isna()
df["Region"] = df.apply(lambda x:"Scotland" if x["Scot"] is True else "Wales" if x["EngReg"]==0 else x["EngReg"],axis=1)
```

```python
df.to_csv("outputs/constituencieswithresultsprocessed.csv", index=False)
```
